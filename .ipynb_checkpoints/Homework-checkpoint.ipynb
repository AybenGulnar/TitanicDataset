```python
#Q9
#As a result of the observations I made with the cross table I prepared, I came to the conclusion that 233 people survived in women, and 
#this number was 109 in men. The total number of surviving passengers is 342. 
#When we look at the survival rates, we come to the conclusion that it is higher in women, with approximately 74.02. 
#I used crosstab to calculate two (or more) factors with a simple crosstab 
df_titanic['Age'] = df_titanic['Age'].fillna((df_titanic['Age'].median()))
t_titanic = pd.crosstab(df_titanic.Sex, df_titanic.Survived, margins=True)
t_titanic['Percentage'] = t_titanic.apply(lambda row: (row[1] / row['All'])*100, axis=1)
t_titanic
#Compute  the  rate  for  each  gender:  number  of survivors for that gender/all genders passengers of the gender
#female: 233/314 = 0,74
#male: 109/577 = 0,18
#female survived / all = 233/891 = 0.26
#male survived/all = 109/891 =0.52
#according table, women's probability of survival is 74%,and while men are 18% below this.
#The probability of survival of women is 1.93320486 times that of the entire population.
#we can also see these rates from the table instead of calculating them one by one. 
#The percentage of men is higher when we compare it to all passengers, this may be due to the fact that men have a higher potential for self-protection and endurance. 
```

```python
#Q10
#I added two missing values to 'Survived S' because of the most count at the S class
#I created a bar chart that shows the number of passengers who survived and who didn't survive for each Port of Embarkation
#The Chart confirms Southampton is the largest for both properties when looking at the number of living and non-living passengers. 
#In addition, the number of passengers who did not live according to the chart is much higher than those who survived. 
def bar_chart(feature):
    survived = df_titanic[df_titanic['Survived']==1][feature].value_counts()
    Unsurvived = df_titanic[df_titanic['Survived']==0][feature].value_counts()
    df = pd.DataFrame([survived,Unsurvived])
    df.index = ['Survived','Unsurvived']
    df.plot(kind='bar', figsize=(7,5),rot=0, colormap='rainbow') #design
    
df_titanic.loc[[61,829],"Embarked"] = "S"
bar_chart('Embarked')
print("Survived :\n",df_titanic[df_titanic['Survived']==1]['Embarked'].value_counts())
print("Unsurvived:\n",df_titanic[df_titanic['Survived']==0]['Embarked'].value_counts())

#Also I used seaborn library
#dfplt = df_titanic.copy(deep=True)
#dfplt["Survived"] = dfplt["Survived"].astype(str)
#print(set(df_titanic["Embarked"].values))
#df_titanic["Embarked"].value_counts()
#sns.catplot(x='Survived', col='Embarked', kind='count',  palette="viridis", data=df_titanic);
#plt.show()
```

```python
#Q11 
#We grouped the ages as adults, children and the olds. 
#I replaced the missing age data with the median of ages 
#If we had written mean instead of median, we would have filled them with mean. In order to get more accurate results, I assigned a value instead of deleting it. 
#Male& survived &agegroup rates:
#adult: 83/109 child: 24/109 old: 2/109
#Female& survived &agegroup rates:
#adult: 184/233 child: 46/233 old: 3/233
#We can find the ratios according to the total number of living people by dividing them by 342. (233+109)
#total unsurvived people = 549
#we can see that the survival rate in adults is quite high compared to other age groups
#The reason why adults have higher survival rates than other categories may be because they can protect themselves better. 
#The children may not have been able to protect it. 
#row[0] = for adult
#row[1] = for child
#row[2] = for old
#Also, if we divide all the categories one by one by the sum of all, we can calculate the rate for all of them. 
#If we compare the ratios, the female unsurvived rate is the highest and the male unsurvived is the least.
#If we look at the numbers, the most surviving female age group is adults. This group is also the highest among male age groups. 
#Since I assigned the median instead of the nan value values, the rate of the median for whichever age group may be higher than normal 
#In the question below, the percentage of child age group was higher among themselves, here the percentage of adults was higher among the groups. 
df_titanic['Age'] = df_titanic['Age'].fillna((df_titanic['Age'].median()))
def agegroup(x):
    if x>=0 and x <=18:
        return 'Child'
    elif x>18 and x<=60:
        return 'Adult'
    else:
        return 'Old'
t_titanic = pd.crosstab([df_titanic.Sex,df_titanic.Survived], df_titanic.Age.apply(agegroup),  margins=True)
t_titanic['Percentage'] = t_titanic.apply(lambda row: (row[1] / row['All'])*100, axis=1)
t_titanic
```

```python
#Q18- #19
# I create a pie chart that shows the percentage of people who survived according to the 
#their age group. For the age group, we will have [0-18] as child, [19-60] as adult, [60 and over] as old. 
#I used the matplotlib library
#I replaced the missing age data with the median of ages 
df_titanic.loc[((df_titanic['Age']>=0) & (df_titanic['Age']<19)) , 'Age Group'] = 'Child'
df_titanic.loc[((df_titanic['Age']>=19) & (df_titanic['Age']<60)) , 'Age Group'] = 'Adult'
df_titanic.loc[(df_titanic['Age']>=60) , 'Age Group'] = 'Old'

#According to my deductions from the chart, 50 percent of the children, 36 percent of the adults and 26 percent of the elderly passengers survived.
#When the survival rates of each age group are examined, it is seen that this rate is highest in children and least in the elderly. 
plt.figure(figsize=(20,5))
plt.subplot(1,3,1)
df_titanic['Survived'][df_titanic['Age Group']=='Child'].value_counts().plot(kind='pie',autopct='%.2f%%',colormap='viridis',
    title = "Child")

plt.subplot(1,3,2)
df_titanic['Survived'][df_titanic['Age Group']=='Adult'].value_counts().plot(kind='pie',autopct='%.2f%%',colormap='viridis',
    title = "Adult")

plt.subplot(1,3,3)
df_titanic['Survived'][df_titanic['Age Group']=='Old'].value_counts().plot(kind='pie',autopct='%.2f%%',colormap='viridis',
    title = "Old")
```
